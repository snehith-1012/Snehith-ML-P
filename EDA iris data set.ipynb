```python
def avg_of_sp(k):
    print(df[df.species==k].petal_length.mean())
    print(df[df.species==k].petal_width.mean())
    print(df[df.species==k].sepal_length.mean())
    print(df[df.species==k].sepal_width.mean())
```

```python
for i in df.species.unique():
    print(i)
    avg_of_sp(i)
    print()
```
